```python
def getAverage(XList):
  return sum(XList)/len(XList)


def getR(XList,YList):

  XAverage=getAverage(XList)
  YAverage=getAverage(YList)

  r=0
  numerator=0 # Числитель
  denSum1=0 # Сумма_1 для знаменателя \summ(X_i-X)**2
  denSum2=0 # Сумма_2 для знаменателя \summ(Y_i-Y)**2
  denominator=0 # Знаменатель

  for i in range (0,100):
    numerator+=(XList[i]-XAverage)*(YList[i]-YAverage)
    denSum1+=(XList[i]-XAverage)**2
    denSum2+=(YList[i]-YAverage)**2
  denominator=math.sqrt(denSum1*denSum2)

  r=numerator/denominator
  return r
```

Проверим гипотезы о независимости признаков (об отсутствии корреляции) с помощью коэффициента корреляции Пирсона.

Пусть $\rho=Corr(X_i,Y_i)$  — истинный (т.е. теоретический, не выборочный) коэффициент корреляции между $(X_i,Y_i)$

### Гипотезы:

$H_0$ - $\rho =0 $ ($X_i$ и $Y_i$ независимы)
 

$H_A$ - $\rho\ne0$ ($X_i$ и $Y_i$ зависимы)

### Условие:

Пусть  $X_i \sim N(\mu_X,\sigma_X^2)$, $Y_i \sim N(\mu_Y,\sigma_Y^2)$, все случайные величины  независимы. Тогда воспользуемся следующей статистикой..

### Статистика:
$$t=\frac{r_{X,Y}\sqrt{n-2}}{\sqrt{1-r_{X,Y}^2}}$$

### Критическое правило:

Основная гипотеза, отвергается, если $|t|>t_{n-2,\frac{\alpha}{2}}$, где  — $\alpha$ уровень значимости

 [$t_{n-2,\frac{\alpha}{2}}$ -- распределение Стьюдента :)].





```python
def getT(r):
  n=100
  return r*math.sqrt(n-2)/(math.sqrt(1-r**2))
```

$H_0$ - $\rho =0 $ ($X_i$ и $Y_i$ независимы)
 

$H_A$ - $\rho\ne0$ ($X_i$ и $Y_i$ зависимы)

### Условие:

Пусть в выборке $(X_1,Y_1),..(X_N,Y_N)$ пары $(X_i,Y_i)$  независимы и одинаково распределены. Тогда воспользуемся следующей статистикой..

### Статистика:
$$t=\frac{r^S_{X,Y}\sqrt{n-2}}{\sqrt{1-(r^S_{X,Y})^2}} $$

### Критическое правило:

Основная гипотеза, отвергается, если $|t|>t_{n-2,\frac{\alpha}{2}}$, где  — $\alpha$ уровень значимости

 [$t_{n-2,\frac{\alpha}{2}}$ -- распределение Стьюдента :)].
